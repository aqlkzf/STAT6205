```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from scipy.stats import norm 

np.random.seed(1)
```

```python
sigma =2

img = pd.read_csv('https://raw.githubusercontent.com/aqlkzf/STAT6205/refs/heads/main/Assignment2/data/letterA.csv',header=None).to_numpy()
mean = np.mean(img)
img2 = np.where(img >= mean, 1, -1)
y = img2 + sigma*np.random.normal(size = img2.shape)

fig, axs = plt.subplots(1,2)
axs[0].imshow(img2, cmap="Greys")
axs[1].imshow(y, cmap="Greys")
# not show the axis
axs[0].axis('off')
axs[1].axis('off')
plt.show()
```
